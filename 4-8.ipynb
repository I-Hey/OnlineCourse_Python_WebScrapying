## ESPN球員球員效率值

[http://insider.espn.com/nba/hollinger/statistics/_/page/1](http://insider.espn.com/nba/hollinger/statistics/_/page/1)

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import random

url = 'http://insider.espn.com/nba/hollinger/statistics/_/page/'

try:
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    chrome = webdriver.Chrome(options=options,
                             executable_path=r'C:\Users\corn_\OneDrive\軟體區\軟體區_程式語言\chromedriver')
    chrome.set_page_load_timeout(10)        # 設定抓取間隔時間
    for i in range(1, 9):           # 8個分頁
        _url = url + str(i)
        print(_url)
        chrome.get(_url)
        soup = BeautifulSoup(chrome.page_source, 'html5lib')
        trs = soup.find('tbody').find_all('tr')
        for tr in trs:
            tds = [td for td in tr.children]        # td = 每欄資料
            rk = tds[0].text.strip()
            if rk =='RK' or len(tds)<2:             # 跳過欄位名稱, 且欄位數小於2 
                continue
            name = tds[1].text
            per = tds[11].text
            print('%s :%s' % (name, per))
        wait = random.randint(2,6)                  # 自動停止抓取時間，偽裝非爬蟲程式用
        print('wait time : %d' % wait)
        time.sleep(wait)
    
    
finally:
    chrome.quit()
```
